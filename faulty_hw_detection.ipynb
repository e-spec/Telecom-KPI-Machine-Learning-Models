```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
# Load telecom KPI dataset from CSV file
data = pd.read_csv('telecom_kpi_data.csv')

# Display the data
print(data)
```

```python
# Features and target
X = data.drop('faulty', axis=1)
y = data['faulty']
#print(X)
#print(y)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
# Train classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

```python
# Evaluation
print("Predictions:", y_pred)
print("Actual:", list(y_test))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## Key Metrics

### 1. Accuracy: 0.9133 (91.33%)

**Formula:**
$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN} = \frac{23 + 114}{23 + 114 + 5 + 8} = \frac{137}{150} = 0.9133$$

**Interpretation:**
- The model correctly classifies **91.33%** of all cases
- Overall, 137 out of 150 predictions are correct
- **Limitation**: Can be misleading with imbalanced data

---

### 2. Precision: 0.8214 (82.14%)

**Formula:**
$$Precision = \frac{TP}{TP + FP} = \frac{23}{23 + 5} = \frac{23}{28} = 0.8214$$

**Interpretation:**
- When the model predicts a cell is faulty, it is correct **82.14%** of the time
- **23 out of 28** positive predictions are correct
- **Use case**: Important when the cost of false alarms is high

---

### 3. Recall (Sensitivity): 0.7419 (74.19%)

**Formula:**
$$Recall = \frac{TP}{TP + FN} = \frac{23}{23 + 8} = \frac{23}{31} = 0.7419$$

**Interpretation:**
- The model correctly identifies **74.19%** of actual faulty cells
- **23 out of 31** actual faults are detected
- **Use case**: Important when missing faults is costly

---

### 4. F1 Score: 0.7797 (77.97%)

**Formula:**
$$F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall} = 2 \times \frac{0.8214 \times 0.7419}{0.8214 + 0.7419} = 0.7797$$

**Interpretation:**
- Harmonic mean of precision and recall
- Balances the trade-off between false positives and false negatives
- **77.97%** overall performance considering both metrics
